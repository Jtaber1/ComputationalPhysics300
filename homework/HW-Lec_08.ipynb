```python
%matplotlib inline  

import matplotlib.pyplot as plt
import numpy as np
```

```python
'''Finding the polynomial to best fit our function f(x)'''

#Parameters
f = lambda x: np.log(x)
a,b = 0.1,10
npoints = 100
order = range(1,11,1)

# Calculate the values at each polynomial degree
for i in order:
    x = np.linspace(a,b,npoints)
    y = f(x) + np.random.rand(npoints) - 0.5
    fit = np.polyfit(x, y, i, full=True)
    print('residual for polynomial of order ' + str(i) + ': ' + str(fit[1]))
```

```python
'''
At the polynomial order of 4, the residuals begin to be within the standard deviation of each other (~ 2.5 units). 
We would only need to use a polynomial of order 4 to fit our function f(x) 
'''
```

```python
'''Graphing the polynomial fit'''

plt.figure(figsize=[12, 8])

# Parameters
f = lambda x: np.log(x)
a,b = 0.1,10
npoints = 100
order = 4

# Generating random points
x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - 0.5

# Curve Fit
fit = np.polyfit(x, y, order, full=True)
y_p = np.polyval(fit[0], x)

# Plotting the function, points, and curve fit. 
plt.plot(x,f(x), label='ideal function')
plt.plot(x,y, 'o', label='noisy data')
plt.plot(x,y_p, label='polyfit'+str(order))

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend(fontsize=12)
plt.show()
```
