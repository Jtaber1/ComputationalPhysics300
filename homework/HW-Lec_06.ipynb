```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def heat_cap(T):
    # Defining variables
    V = .001    # m^3 
    k_B = 1.38 * 10**(-23)    # Js/K
    theta = 428    # K
    density = 2710   # kg/m^3
    
    # Defining the integral function
    def f(x):
            return x**4 * np.exp(x) / (np.exp(x)-1)**2
    
    # b is an arrary so we have to integrate for each value of T 
    integrals = []
    for i in T: 
        I = integrate.quad(f, 0, theta/i)      # outputs [area, error]
        integral = I[0]                        # Only need the area value
        integrals.append(integral)
        
    return 9 * V * density * k_B * (T/theta)**3 * integrals

T = np.linspace(5, 500, 1000)
y = heat_cap(T)
plt.plot(T, y)
plt.xlabel('T (K)')
plt.ylabel('$C_V$ (J/kg K)')
plt.show()
```
