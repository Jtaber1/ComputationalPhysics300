```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''Rectangle Rule'''

def Rectangle(f, a, b, parts, plot=1):
    # Defining the X,Y points
    deltax = (b - a) / parts      # Width of the partition
    resultsx = np.linspace(a, b, parts)      # Returns evenly spaced numbers over the interval (parts)
    resultsy = f(resultsx)
    
    # Area Calculation
    area = np.empty([parts], float)
    for i in range(parts):
            area[i] = resultsy[i] * deltax 
    

    # Graphing 
    if plot==1:
        x = np.linspace(a,b,100)
        y = f(x)    
        plt.plot(x, y, 'r')
        plt.bar(resultsx+deltax/2, resultsy, deltax, edgecolor ='black')      
        plt.show()
        print("The Sum of the area is: ", sum(area))
    
    # Adding the area of each bar together
    return sum(area)
```

```python
'''Integrals using the rectangle method'''

Rectangle(lambda x: x * x  - x + 1, 0, 3, 10)
Rectangle(lambda x: x**4 - x + 1, 0, 3, 10)
Rectangle(lambda x: np.exp(-(x**2)), 0, 3, 10)
```

```python
'''Trapezoidal Rule '''

def Trapezoid(f, a, b, parts, plot=1):
    #define the X, Y points 
    deltax = (b - a) / parts
    resultsx = np.linspace(a, b, parts+1)
    resultsy = f(resultsx)

    # Area calculation
    area = np.empty([parts], float)
    for i in range(parts):
            area[i] = (resultsy[i]+resultsy[i+1]) * deltax/2      # Area of the bar added to the area of the triangle

    # Graphing
    if plot==1:
        x = np.linspace(a, b,100)
        y = f(x)        
        plt.plot(x, y, 'r')
        plt.xlim([a,b])
        
        y2 = np.array([0,0])
        for i in range(parts):
            x0 = resultsx[i:i+2]
            y1 = resultsy[i:i+2]
            plt.fill_between(x0, y1, y2, where=y1>=y2, facecolor='blue')
            linex, liney = [resultsx[i+1], resultsx[i+1]], [0, resultsy[i+1]]
            plt.plot(linex, liney, color='black', linewidth=2.0)        
        
        plt.show()
        print("The Sum of the area is: ", sum(area))
    
    return sum(area)
```

```python
'''Integrals using the trapezoid method'''

Trapezoid(lambda x: x * x  - x + 1, 0, 3, 10)
Trapezoid(lambda x: x**4 - x + 1, 0, 3, 10)
Trapezoid(lambda x: np.exp(-(x**2)), 0, 3, 10)
```

```python
'''Simpson's Rule'''

def Simpson(f, a, b, parts, plot=1): 
    deltax = (b - a) / parts
    
    # Defining the area using Simpson's composite rule (allows for more partitions)
    sum_1 = 0
    sum_2 = 0
    # First defining the two alternating summations (1/3, 4/3, 2/3, ..., 4/3,1/3) 
    for i in range(1, parts, 2):      # Odd summations (coefficient of 4)
        sum_1 += f(a + i * deltax)      # x_j = a + ih, h = deltax
    for i in range(2, parts - 1, 2):      # Even summations (coefficient of 2)
        sum_2 += f(a + i * deltax)
    
    area = 1/3 * deltax * (f(a) + 4*sum_1 + 2*sum_2 + f(b))
    
    
    #Graphing
    if plot==1:
        x = np.linspace(a, b, 100)
        y = f(x)        
        plt.plot(x, y, 'r')  
        plt.plot()
        plt.xlim([a,b])
        plt.ylim([min(y),max(y)])
        plt.show()
        print("The Sum of the area is: ", area)
        
    return area
```

```python
'''Integrals using the Simpson's rule'''

Simpson(lambda x: x*x - x + 1, 0, 3, 2)
Simpson(lambda x: x**4 - x + 1, 0, 3, 8)
Simpson(lambda x: np.exp(-(x**2)), 0, 3, 6)
```
