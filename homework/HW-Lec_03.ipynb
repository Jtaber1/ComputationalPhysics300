```python
'''
For this assignment we will be graphing the energy values of two interacting Einstein solids 
against the Muliplicity of the system to find the most likely energy state. We will do this by 
creating an empty array and calculating the total multiplicty of the system for each combination 
of energy states.
I followed the code given from lecture III in this assignment to generate the graphs.  
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import log, exp
```

```python
'''
We first define the function we will use to calculate the multiplicity of each Einstein solid in
the system. For large systems, python cannot compute large factorials, so we use the exp(log(x)) = x
trick so we can deal with smaller numbers. 
Also used is the property log(a/b) = log(a) - log(b)
In this program I will use numpy's factorial function instead of the one we used in lecture II and 
for homework II.
'''

# Factorial function using numpy's built in factorial function
def fact(x):
    return np.math.factorial(x)


# Multiplicity function using the log trick due to the large numbers involved
def Multiplicity(N, q):
    return exp(log(fact(q + N - 1))-log(fact(q))-log(fact(N - 1)))
```

```python
'''
To generate a graph of the total multiplicty we will first create an empty list where we can store calculated
values of the total multiplicity. As we calculate each value for total multiplicty based on the combination of 
qa and qb, we appened the total multiplicity onto this list. We can then use this list and graph it against the
given possible values of qa. 
'''

# Defining the parameters of the system
Na = 300
Nb = 200
q_total = 100

qa_list = range(q_total)    # Creating a list of possible values of qa between the range of the total energy (0-100)
                            # qb = q_total - qa
omega_total = []     # Ceating an empty list for multiplicity that will be appended with calculated values

# For every possible value of qa, we want to calculate the total multiplicty
for qa in qa_list:
    qb = q_total - qa       #total energy is the sum of all energies
    omega_a = Multiplicity(Na, qa)
    omega_b = Multiplicity(Nb, qb)
    omega_total.append(omega_a * omega_b)     # each calculated value is added and stored to the list
    
plt.plot(qa_list, omega_total)      # qa_list and not qa since qa_list includes all posible values of qa
plt.xlabel('$q_a$')
plt.ylabel('$\Omega$')
plt.show()
```

```python
# Changing the parameters
Na = 3000
Nb = 2000
q_total = 100

qa_list = range(q_total)   
omega_total = []     


for qa in qa_list:
    qb = q_total - qa      
    omega_a = Multiplicity(Na, qa)
    omega_b = Multiplicity(Nb, qb)
    omega_total.append(omega_a * omega_b)     
    
plt.plot(qa_list, omega_total)      
plt.xlabel('$q_a$')
plt.ylabel('$\Omega$')
plt.show()
```
